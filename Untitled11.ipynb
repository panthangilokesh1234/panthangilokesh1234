```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a_input, b_input, c_input):
    # Convert inputs to float
    a = float(a_input)
    b = float(b_input)
    c = float(c_input)

    # Calculate temperature using quadratic equation
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    # Get user inputs for coefficients a, b, and c
    for i in range(3):
      a_input = input("Enter the value for a: ")
      b_input = input("Enter the value for b: ")
      c_input = input("Enter the value for c: ")

      # Generate time values from 0 to 10 with a step of 0.1
      time = np.arange(0, 10, 0.1)

      # Calculate temperatures using user-provided coefficients
      temperature_keyboard = quadratic_model(time, a_input, b_input, c_input)

      # Plotting
      plt.plot(time, temperature_keyboard, label='user input coefficients')
    plt.xlabel('time')
    plt.ylabel('temperature')
    plt.legend()
    plt.title('Weather modeling with quadratic equation with user input')
    plt.show()

if __name__ == "__main__":
    main()
```
